```python
import feather
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.ensemble import RandomForestRegressor
from sklearn.datasets import make_regression
from sklearn.model_selection import train_test_split
from scipy import stats 

#This file is trying to figure out the best number of trees to use. 
#I'll just compare the R2 from a bunch of different runs of TrainSet1 
#with a few different n_estimator values.
```

```python
metadata_Train1 = feather.read_dataframe('../out/GaussianProcessModeling/in/metadata_Train1_all.feather')
metadata_Test1 = feather.read_dataframe('../out/GaussianProcessModeling/in/metadata_Test1_all.feather')
features_Train1 = feather.read_dataframe('../out/GaussianProcessModeling/in/features_Train1_all.feather')
features_Test1 = feather.read_dataframe('../out/GaussianProcessModeling/in/features_Test1_all.feather')
#feature data already organized by DADA_id, but the index column is missing so need to keep 
#that in mind when importing the larger dataset (changed in R so that index is DADA_id)

ages_Train1=metadata_Train1.iloc[:,2]

ages_Test1=metadata_Test1.iloc[:,2]

#Prep the regressor
regressor = RandomForestRegressor(n_estimators=100, random_state=0)
# fit it with the training data
regressor.fit(features_Train1, ages_Train1)
# make predictions on the test input
predicted_Test1 = regressor.predict(features_Test1)

# plot our true test Y's to our predicted Y's
plt.figure(figsize=(8, 8))
plt.scatter(ages_Test1, predicted_Test1, s=1)
#plt.xlim(0, 30)
#plt.ylim(0, 30)
plt.ylabel('Predicted Age')
plt.xlabel('Chronological Age')
plt.title('Full Dataset, 80:20 split, Test Set 1')

metadata_Test1['age_mb']=predicted_Test1
#feather.write_dataframe(metadata_Test1, '../out/RandomForests/out/All_PredictedAgeData_Test1.feather')

gradient, intercept, r_value, p_value, std_err =stats.linregress(predicted_Test1,ages_Test1)
print("Slope",gradient)
print("Intercept",intercept)
print("p-value",p_value)
print("R-squared",r_value**2)
print("Pearson's R, p-value",stats.pearsonr(predicted_Test1,ages_Test1))
```
