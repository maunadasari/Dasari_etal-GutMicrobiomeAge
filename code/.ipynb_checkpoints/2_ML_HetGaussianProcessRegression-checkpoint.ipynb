```python
#Load my packages
import feather
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.gaussian_process import GaussianProcessRegressor
from sklearn.gaussian_process.kernels import RBF
from scipy import stats 

#Because I'm terrible at python and afraid of messing up the 
#matching of training and test data, I'm going to do it in R
```

```python
metadata_Train1 = feather.read_dataframe('../data/metadata_Train1_all.feather')
metadata_Test1 = feather.read_dataframe('../data/metadata_Test1_all.feather')
features_Train1 = feather.read_dataframe('../data/features_Train1_all.feather')
features_Test1 = feather.read_dataframe('../data/features_Test1_all.feather')
#feature data already organized by DADA_id, but the index column is missing so need to keep 
#that in mind when importing the larger dataset (changed in R so that index is DADA_id)

ages_Train1=metadata_Train1.iloc[:,2]

ages_Test1=metadata_Test1.iloc[:,2]

#define the kernel
l_scale = 34.38
l_bounds = 1, 100
kernel = np.var(ages_Train1)*RBF(length_scale=l_scale, length_scale_bounds=l_bounds)

# instantiate the Gaussian process
gp = GaussianProcessRegressor(kernel=kernel)
# fit it with the training data
fit = gp.fit(features_Train1, ages_Train1)
# make predictions on the test input
predicted_Test1 = fit.predict(features_Test1)

# plot our true test Y's to our predicted Y's
plt.figure(figsize=(8, 8))
plt.scatter(ages_Test1, predicted_Test1, s=1)
#plt.xlim(0, 30)
#plt.ylim(0, 30)
plt.ylabel('Predicted Age')
plt.xlabel('Chronological Age')
plt.title('Full Dataset, 80:20 split, Test Set 1')

metadata_Test1['age_mb']=predicted_Test1
feather.write_dataframe(metadata_Test1, '../data//H.All_PredictedAgeData_Test1.feather')

gradient, intercept, r_value, p_value, std_err =stats.linregress(predicted_Test1,ages_Test1)
print("Slope",gradient)
print("Intercept",intercept)
print("p-value",p_value)
print("R-squared",r_value**2)
print("Pearson's R, p-value",stats.pearsonr(predicted_Test1,ages_Test1))
```
